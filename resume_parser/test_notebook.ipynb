```python
from spacy.tokens import Token
valid_name_type=['Borrower','Seller','Lender','Contractor']
def is_valid_type(token):
    for word in valid_name_type:
            if word in token.text:
                return True
    else:
        return False 
def check_next(token):
    try:
        temp=token.nbor()
        if temp.pos_ in ['NOUN','PROPN'] and is_valid_type(temp):
            return True
        elif temp.pos_ =='PUNCT':
            return check_next(temp)
        else:
            return  False
    except:
        return False

Token.set_extension("is_valid_type", getter=is_valid_type,force=True)
Token.set_extension("check_ner", getter=lambda token: token.ent_type_ in ["ORG","PERSON"],force=True)
Token.set_extension("validate_pattern_1",getter=lambda token: token.pos_ in ["NOUN","PROPN"],force=True)
Token.set_extension("check_next", getter=check_next,force=True)
```
